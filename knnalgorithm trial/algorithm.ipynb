```python
# imorting the required modules 
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from statistics import mode
```

```python
# loading the data set 
dataset = pd.read_csv("Employee.csv")
```

```python
# to see the first five rows of the data set 
dataset.head()
```

```python
# seperating dependent and independent variables 
X = dataset[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
Y = dataset[['Outcome']]
# converting them into the array to load it as test and train data set  
xnew = X.to_numpy()
ynew = Y.to_numpy()
```

```python
#  importing sklearn module and from it importing th knn-classifier   the method to split the dataset into test ans train  
from sklearn.neighbors import KNeighborsClassifier
# importing sklearn module and the method to split the dataset into test ans train
from sklearn.model_selection import train_test_split
```

```python
# spliting the data set to train and test data 
X_train, X_test, Y_train, Y_test = train_test_split(xnew, ynew, test_size = 0.20, random_state=0)
```

```python
# the mian block for the algorithm to work on 
final_list = []
test_list = [ [i] for i in X_test]
# this loop gives the test data one by one to the algorithm 
for i in range(len(test_list)) :
    # this loop assumes many k values to build the model 
    for k in range(7,39,2):
        # creating a model for each neighbour value
        model = KNeighborsClassifier(n_neighbors=k)
        # fitiign the model with th train data 
        model.fit(X_train,Y_train)

        if k == 7:
            y = list(model.predict(test_list[i]))
        else:
            y += list(model.predict(test_list[i]))
    # gives the highest frequency  value from all the values of k  
    final_list.append(mode(y))
```

```python
y = final_list
#  importing a supported method to find the acuuracy of teh model 
from sklearn.metrics import accuracy_score
ac1 = accuracy_score(Y_test,y)
print(ac1)
```

```python
# this the normal form of knn-classifier
model = KNeighborsClassifier(n_neighbors=33)
model.fit(X_train,Y_train)
y_pred = model.predict(X_test)
ac = accuracy_score(Y_test,y_pred)
print(ac)


```
